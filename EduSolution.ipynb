```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt



data_train=pd.read_csv("trainu.csv")
data_train.head()
plt.figure(figsize=(13,7))
plt.bar(data_train["Co.Name"],data_train["quality"])
plt.title("Graph of quality of education")
plt.xlabel("college names")
plt.ylabel("quality of education provided by the education system")
plt.show()
print()

plt.figure(figsize=(13,7))
plt.plot(data_train["Co.Name"],data_train["Pass out"])
plt.title("Graph of Pass out")
plt.xlabel("college names")
plt.ylabel("Percentage of pass out of students")
plt.show()
print()

plt.figure(figsize=(13,7))
plt.plot(data_train["Co.Name"],data_train["infra:%"])
plt.title("Graph of Availabilty of Infrastructure")
plt.xlabel("college names")
plt.ylabel("Availabilty of infrastructure")
plt.show()


                                                                                                                                              
```

```python
Y_train = pd.DataFrame(data = data_train.iloc[:, 1].values, columns = ['quality'])

data_train.head()     
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_train, Y_train, test_size=0.1, random_state=42)
print(X_test)

# X_train.shape  // to see the size of the dataset


# Linear Regression
from sklearn import linear_model
linear_regression = linear_model.LinearRegression()
linear_regression.fit(X_train, y_train)
#Saving Scikitlearn models
import joblib
joblib.dump(linear_regression, "predictquality.pkl")
from sklearn.model_selection import train_test_split
model=joblib.load("predictquality.pkl")
# test cases
testresult=model.fit(X_test,y_test)
```

```python
# lets predict some quality out comes of few college with fewer data with testu data set
# lets import the file and  delete some colums which we do not needed 
data_test=pd.read_csv("testu.csv")
data_test=data_test.drop("Co.Name",axis=1)
data_test=data_test.drop("Total Strg",axis=1)
data_test.head()
result=model.predict(data_test)
print(result)
```
